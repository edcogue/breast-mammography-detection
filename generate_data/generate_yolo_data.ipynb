```python
# We import all our dependencies.
import numpy as np
import sys
import os
sys.path.append('/tf/code/preprocess/data_loader')
from data_loader import DataLoader
from collections import defaultdict
import cv2
```

```python
def generate_yolo_box(bboxes, image_shape):
    bboxes = np.array(bboxes)
    #Calc the xcenter and ycenter. bboxes = [xmin, ymin, w, d]
    xcenter = bboxes[0] + bboxes[2]/2
    ycenter = bboxes[1] + bboxes[3]/2
    
    #Calc proportion to image
    xcenter = xcenter / image_shape[1]
    ycenter = ycenter / image_shape[0]
    w = bboxes[2] / image_shape[1]
    h = bboxes[3] / image_shape[0]
    return xcenter, ycenter, w, h   
```

```python
loader = DataLoader()
```

```python
# generate YOLO data
DETECTION_YOLO_BASE_DIR="/kaggle/working/yolo_data/object_detection"
def get_yolo_detection_dir(type_set, label_or_image):
    assert type_set in ["train", "val", "test"]
    assert label_or_image in ["labels", "images"]
    return os.path.join(DETECTION_YOLO_BASE_DIR, type_set, label_or_image)

for type_set in ["train", "val", "test"]:
    for type_data in ["labels","images"]:
        os.makedirs(get_yolo_detection_dir(type_set, type_data), exist_ok=True)
```

```python
od_generator= loader.object_detection_generator(return_filename=True, denoise=True)
for img, boxes, file in od_generator():
    type_set = np.random.choice(["train", "val", "test"], p=[0.7, 0.2, 0.1])
    
    if not boxes:
        continue

    with open(os.path.join(get_yolo_detection_dir(type_set, "labels"), file+".txt"),"w") as f:
        for box in boxes:
            yolo_box = generate_yolo_box(box, img.shape[:2])
            f.write(f"0 {yolo_box[0]} {yolo_box[1]} {yolo_box[2]} {yolo_box[3]} \n")

    cv2.imwrite(f"{os.path.join(get_yolo_detection_dir(type_set, 'images'), file+'.png')}", img)
```

```python
CLASS_YOLO_BASE_DIR="/kaggle/working/yolo_data/classification"
def get_class_yolo_dir(type_set, class_label):
    assert type_set in ["train", "val", "test"]
    return os.path.join(CLASS_YOLO_BASE_DIR, type_set, class_label)
```

```python
loader = DataLoader()

counter_labels = defaultdict(int)
class_generator= loader.classification_generator(output_size=(300,300))
counter=0
for image, label in class_generator():
    type_set = np.random.choice(["train", "val", "test"], p=[0.7, 0.2, 0.1])
    os.makedirs(get_class_yolo_dir(type_set, label), exist_ok=True)
    counter_labels[label] +=1
    cv2.imwrite(f"{os.path.join(get_class_yolo_dir(type_set, label), str(counter_labels[label])+'.png')}", image)
    print(counter, end="\r", flush=True)
    counter+=1
```
